```python
import pandas as pd
import os
import requests
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
EPL_df = pd.read_csv("../final-resources/EPL_scores_att_weather.csv")
Bundesliga_df = pd.read_csv("../final-resources/Bundesliga_scores_weather.csv")
LaLiga_df =  pd.read_csv("../final-resources/la_liga_scores_weather.csv")
SerieA_df = pd.read_csv("../final-resources/Serie_A_weather.csv")
```

```python
EPL_df.head()
# Create bins in which to place values based upon TED Talk views
bins = [0, 39, 49, 59, 69, 79, 89]

# Create labels for these bins
group_labels = ["0 to 40", "40 to 50", "50 to 60", "60 to 70", "70 to 80", "80 to 90"]

#put data in bins
EPL_df["Temperature Group"]=pd.cut(EPL_df["Avg Temp (in F)"], bins, labels=group_labels)
# Create a GroupBy object based upon "View Group"
EPL_dfgroup = EPL_df.groupby("Temperature Group")
home = EPL_dfgroup["HomeScore"].mean()
away = EPL_dfgroup["AwayScore"].mean()
fig,ax = plt.subplots(figsize=(10,5))
n = len(home)
x = np.arange(n)
w = 0.4
ax.set_title(f"EPL Comparison of Average Number of Goals in Temperature Ranges (F) ", fontsize=20)
ax.bar(x-w/2, home, color ="crimson", edgecolor ="gray", label ="Home Score", width = w)
ax.bar(x+w/2, away, color ="midnightblue", edgecolor ="gray", label ="Away Score", width = w)
#  ax.set_xlim(-0.5, n-0.5)
ax.set_xticks(x)
ax.set_xticklabels(group_labels)
ax.set_ylabel("Number of wins", fontsize=10)
ax.set_xlabel("Temperature Ranges in F", fontsize = 10)
ax.legend()
fig.savefig(f"EPL_weather_bins.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```
