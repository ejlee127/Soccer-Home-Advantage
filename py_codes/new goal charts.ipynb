```python
import pandas as pd
import matplotlib
from matplotlib import pyplot
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
import os
import numpy as np
```

```python
# all the csvs
England_pre_df = pd.read_csv("england_pre.csv")
England_post_df = pd.read_csv("england_post.csv")
Germany_pre_df = pd.read_csv("germany_pre.csv")
Germany_post_df = pd.read_csv("germany_post.csv")
Italy_pre_df = pd.read_csv("Italy_pre.csv")
Italy_post_df = pd.read_csv("Italy_post.csv")
Spain_pre_df = pd.read_csv("spain_pre.csv")
Spain_post_df = pd.read_csv("spain_post.csv")
```

```python
# make Eng bar plot
# get values of the six bars
e1 = England_pre_df["HomeScore"].mean()
e2 = England_post_df["HomeScore"].mean()
e3 = England_pre_df["AwayScore"].mean()
e4 = England_post_df["AwayScore"].mean()
e5 = e1+e3
e6 = e2+e4
```

```python
# group the bars
ehome = [e1, e2]
eaway = [e3, e4]
etotal = [e5, e6]

# set width of bars
barWidth = 1
plt.subplots(figsize =(12,10))

ebar1 = [0,4]
ebar2 = [1,5]
ebar3 = [2,6]
```

```python
Eng1 = plt.bar(ebar1, ehome, color ="crimson", width = barWidth, edgecolor ="gray", label ="Home")
Eng2 = plt.bar(ebar2, eaway, color ="mistyrose", width = barWidth, edgecolor ="gray", label ="Away")
Eng3 = plt.bar(ebar3, etotal, color ="royalblue", width = barWidth, edgecolor ="gray", label ="Total")

plt.ylabel("Goals scored per match")
plt.xlabel("With fans           |           Without fans")
labels = ["1.51", "1.22", "2.72", "1.58", "1.11", "2.69"]
plt.xticks([0, 1, 2, 4, 5, 6], labels)
plt.title("Goals per match with and without fans - England")
plt.legend()

plt.savefig("EPL4.png")
plt.show()
```
