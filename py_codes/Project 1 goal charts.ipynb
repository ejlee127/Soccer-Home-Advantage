```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show
import os
import numpy as np
```

```python
# all the csvs
England_pre_df = pd.read_csv("england_pre.csv")
England_post_df = pd.read_csv("england_post.csv")
Germany_pre_df = pd.read_csv("germany_pre.csv")
Germany_post_df = pd.read_csv("germany_post.csv")
Italy_pre_df = pd.read_csv("Italy_pre.csv")
Italy_post_df = pd.read_csv("Italy_post.csv")
Spain_pre_df = pd.read_csv("spain_pre.csv")
Spain_post_df = pd.read_csv("spain_post.csv")
```

```python
# make Eng bar plot
# get values of the six bars
e1 = England_pre_df["HomeScore"].mean()
e2 = England_post_df["HomeScore"].mean()
e3 = England_pre_df["AwayScore"].mean()
e4 = England_post_df["AwayScore"].mean()
e5 = e1+e3
e6 = e2+e4
```

```python
# group the bars
epre = [e1, e3, e5]
epost = [e2, e4, e6]
labels = ["Home", "Away", "Total"]

# set width of bars
width = 1
plt.subplots(figsize =(12,10))

ebar1 = [0,3,6]
ebar2 = [1,4,7]
```

```python
Eng1 = plt.bar(ebar1, epre, color ="crimson", width = barWidth, edgecolor ="gray", label ="With fans")
Eng2 = plt.bar(ebar2, epost, color ="mistyrose", width = barWidth, edgecolor ="gray", label ="Without fans")

plt.ylabel("Goals scored per match")
plt.title("Goals per match with and without fans - England")
plt.xticks([0.5,3.5,6.5], ["Home", "Away", "Total"])
plt.legend()

# print means of 6 categories
print ("Home teams averaged 1.507 goals per match with fans in attendance.")
print ("Home teams averaged 1.581 goals per match without fans in attendance.")
print ("Away teams averaged 1.215 goals per match with fans in attendance.")
print ("Away teams averaged 1.113 goals per match without fans in attendance.")
print ("Matches averaged a total of 2.722 goals per match with fans in attendance.")
print ("Matches averaged a total of 2.694 goals per match without fans in attendance.")

plt.savefig("EPL3.png")
plt.show()
```
