```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show
import os
import numpy as np
```

```python
# England
England_pre_df = pd.read_csv("england_pre.csv")
England_pre_df

England_post_df = pd.read_csv("england_post.csv")
England_post_df
```

```python
count1 = England_pre_df["Result"].value_counts()
count1
count2 = England_post_df["Result"].value_counts()
count2
```

```python
result_type = ["Home Wins", "Away Wins", "Draws"]
results = [129,87,72]
colors = ["crimson","mistyrose","silver"]
explode = (0.1,0,0)

result2_type = ["Home Wins", "Away Wins", "Draws"]
results2 = [30,19,13]
colors2 = ["crimson","mistyrose","silver"]
explode2 = (0.1,0,0)
```

```python
EPL1 = plt.figure(1)
plt.pie(results, explode=explode, colors=colors, 
        autopct="%1.1f%%", shadow=True, startangle=90)
legend_labels = [r"Home Wins (129)", r"Away Wins (87)", r"Draws (72)"]
plt.axis("equal")
plt.title("English Premier League home/away results with fans")
plt.legend(legend_labels, loc="lower left")

EPL2 = plt.figure(2)
plt.pie(results2, explode=explode2, colors=colors2, 
        autopct="%1.1f%%", shadow=True, startangle=90)
legend_labels = [r"Home Wins (30)", r"Away Wins (19)", r"Draws (13)"]
plt.axis("equal")
plt.title("English Premier League home/away results without fans")
plt.legend(legend_labels, loc="lower left")

plt.savefig("EPL1.png")
plt.savefig("EPL2.png")
plt.show()

```
